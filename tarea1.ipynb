```
import pandas as pd
import matplotlib.pyplot as plt

githubURL = "https://raw.githubusercontent.com/mexcoder/ReconocimientoPatrones/master/Datasets/DataSet1.csv"

# cargar los datos desde github
df = pd.read_csv(githubURL)

# reordenar los valores para que x quede en orden ascendente
df = df.sort_values("x")
# comprobar los datos
print (df)
print("\r\n\r\n")
plot = df.plot.scatter(x="x", y="y")
plt.show()
```

# Regresion Lineal
Para la regresion se utilizara la columna "x" como la variable dependiente

```
import numpy as np
from sklearn.linear_model import LinearRegression

# convertir el dataset de pandas a un arreglo de numpy
x = np.asarray(df[['x']])
y = np.asarray(df[['y']])

# arreglo extra para probar el modelo
x_new = np.linspace(0, 1,100).reshape(100,1)

# instanciar el modelo
model = LinearRegression()

# entrenar el modelo
model.fit(x,y)

# tratar de predecir el arreglo original
y_pred = model.predict(x)
# tratar de predecir el arreglo nuevo
y_pred2 = model.predict(x_new)

# graficar todo
plt.scatter(x, y)
plt.plot(x,y_pred,'--r')
plt.plot(x,y_pred2,'--y')
plt.show()

# calcular el coeficiente R^2
print ()
print ("Coeficiente R²:", model.score(X=x, y=y))
```

# KNN
para knn igual se utilizara X como variable independiente pero para determinar el mejor valor para "K" se iterara tomando encuenta las restricciones del modelo

```
from sklearn.neighbors import KNeighborsRegressor

# convertir el dataset de pandas a un arreglo de numpy
x = np.asarray(df[['x']])
y = np.asarray(df[['y']])

# iterar k para intentar encontrar la mejor opcion
for k in range (3, 25, 2):
  # instanciar el modelo
  knn = KNeighborsRegressor(n_neighbors=k)

  # entrenar el modelo
  knn.fit(x,y)

  # tratar de predecir el arreglo original
  y_pred = knn.predict(x)
  
  # imprimir el valor actual de "K"
  print(k)

  #graficar
  plt.scatter(x,y)
  plt.plot(x,y_pred,'--y')
  plt.show()

  # calcular el coeficiente R^2
  print ("Coeficiente R²:", model.score(x, y))
```

Despues de comparar las graficas decidi elegir K=11 ya que muestra una buena similutud a los datos originales sin perder los bordes debido a la falta de vecinos

```
# valor elegido para k
k = 11

# arreglo extra para probar el modelo
x_new = np.linspace(0, 1,100).reshape(100,1)

# instanciar el modelo
knn = KNeighborsRegressor(n_neighbors=k)

# entrenar el modelo
knn.fit(x,y)

# tratar de predecir el arreglo original
y_pred = knn.predict(x) 
y_pred2 = knn.predict(x_new)

#graficar
plt.scatter(x,y)
plt.plot(x,y_pred,'--y')
plt.plot(x,y_pred2,'--g')
plt.show()

# calcular el coeficiente R^2
print ("Coeficiente R²:", model.score(x, y))
```
